```python
import pandas as pd

agri = pd.read_csv('/content/Table_6.2_Status_of_farm_machinery_industries_in_India.csv')
climate = pd.read_csv('/content/seasonalFrequency_cd_Winter-1891-2021.csv')

print("Agriculture data shape:", agri.shape)
print("Climate data shape:", climate.shape)

agri.head(), climate.head()
```

```python
agri.columns = [c.strip().replace(' ', '_') for c in agri.columns]
climate.columns = [c.strip().replace(' ', '_') for c in climate.columns]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(climate['Year'], climate['January:_TOTAL'], label='Jan_Total')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Winter Season Frequency Over Time (1891â€“2021)')
plt.legend()
plt.show()
```
